# **CONFIDENCE INTERVAL (CI)**

We then convert **normal distributed** to **standard normal distributed** form:

```
x = ideal_normal
x_s = (x - x.mean()) / x.std()

sns.distplot(ideal_normal, color='orange').set_title('Normal Distribution')
plt.show()
sns.distplot(x_s, color='orange').set_title('Standard Normal Distribution')
plt.show()
```

After Converted to **Standard Normal Distribution**, We now can analyse the **Confidence Interval (CI)**

CI at 90% Siginificance level

```
z_interval = stats.norm.interval(alpha=0.90)
z_interval
```
