```
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests

print("Mean of data = ", df.Energy.mean())
print("Standard deviation of data = ", df.Energy.std())

ztest ,pval = stests.ztest(df.Energy, x2=None, value=90)
print("Value of Z_test = ", ztest)

z_interval = stats.norm.interval(alpha=0.99)
print("Z_interval = ",z_interval)

if pval<0.01:
    print("since Z_test < -Z_interval or Z_test > Z_interval, we reject null hypothesis")
else:
    print("since -Z_interval < Z_test > Z_interval, we accept null hypothesis")
```
